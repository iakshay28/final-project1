```python
class FoodItem:
    next_id = 1

    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = FoodItem.next_id
        FoodItem.next_id += 1
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

    def __str__(self):
        return f"FoodID: {self.food_id}, Name: {self.name}, Quantity: {self.quantity}, Price: {self.price}, Discount: {self.discount}, Stock: {self.stock}"

class FoodMenu:
    def __init__(self):
        self.menu = []

    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        self.menu.append(food_item)

    def edit_food_item(self, food_id, name=None, quantity=None, price=None, discount=None, stock=None):
        for food_item in self.menu:
            if food_item.food_id == food_id:
                if name is not None:
                    food_item.name = name
                if quantity is not None:
                    food_item.quantity = quantity
                if price is not None:
                    food_item.price = price
                if discount is not None:
                    food_item.discount = discount
                if stock is not None:
                    food_item.stock = stock
                break

    def view_all_food_items(self):
        return self.menu

    def remove_food_item(self, food_id):
        self.menu = [food_item for food_item in self.menu if food_item.food_id != food_id]
        
food_menu = FoodMenu()

food_menu.add_food_item("Tandoori Chicken", "4 pices", 240, 0, 10)
food_menu.add_food_item("Vegan Burger", "1 pices", 320, 0, 5)
food_menu.add_food_item("Truffle Cake", "500 gm", 900, 0, 20)

print([str(food_item) for food_item in food_menu.view_all_food_items()])

food_menu.edit_food_item(2, discount=1.0)
print([str(food_item) for food_item in food_menu.view_all_food_items()])


class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.order_history = []
        
    def login(self, email, password):
        if self.email == email and self.password == password:
            return True
        else:
            return False
        
    def update_profile(self, full_name=None, phone_number=None, address=None, password=None):
        if full_name:
            self.full_name = full_name
        if phone_number:
            self.phone_number = phone_number
        if address:
            self.address = address
        if password:
            self.password = password
            
    def place_order(self, food_list):
        order = []
        food_items = [("Tandoori Chicken", 240, 4),
                      ("Vegan Burger", 320, 1),
                      ("Truffle Cake", 900, 500)
                     ]
        for food_item in food_list:
            order.append(food_items[food_item - 1])
        self.order_history.append(order)
        return order
    
    def view_order_history(self):
        return self.order_history

user = User("akshay kumar", "8985454894", "akshay.inamdar1995@gmail.com", "8-25", "password")


if user.login("akshay.inamdar1995@gmail.com", "password"):
    while True:
        print("1. Place New Order")
        print("2. Order History")
        print("3. Update Profile")
        option = int(input("Enter an option: "))

        if option == 1:
            print("1. Tandoori Chicken (4 pieces) [INR 240]")
            print("2. Vegan Burger (1 Piece) [INR 320]")
            print("3. Truffle Cake (500gm) [INR 900]")
            food_list = [int(x) for x in input("Enter the food items you want to order (separated by space): ").split()]
            order = user.place_order(food_list)
            print("Order: ", order)
            
        elif option == 2:
            order_history = user.view_order_history()
            print("Order History: ", order_history)
            
        elif option == 3:
            full_name = input("Enter Full Name: ")
            phone_number = input("Enter Phone Number: ")
            address = input("Enter Address: ")
            password = input("Enter Password: ")
            user.update_profile(full_name, phone_number, address, password)
            
        else:
            break
else:
    print("Login failed.")
    

```
